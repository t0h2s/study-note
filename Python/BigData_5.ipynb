```python
import pandas as pd
import numpy as np

df = pd.read_csv('./datasets/bookings.csv')

df['Total_Review'] = df['Total_Review'].map(lambda x: str(x).replace('external','').strip())
df['Total_Review'] = df['Total_Review'].map(lambda x: str(x).replace('review','').strip())
df['Total_Review'] = df['Total_Review'].map(lambda x: str(x).replace(',',''))
df['Total_Review'] = df['Total_Review'].astype('float') # 데이터 타입 변환
df['Total_Review'].describe(include='int') # 수치형 변수들에 대한 요약 통계량 확인
```

```python
# isna()
# 값이 결측치인지 확인

import pandas as pd
import numpy as np

df = pd.read_csv('./datasets/bookings.csv')

index = df[df['Rating'].isna()].head(5).index
index
```

```python
# fillna()
# 결측치를 인자에 전달하는 값으로 대체

import pandas as pd
import numpy as np

df = pd.read_csv('./datasets/bookings.csv')

df['Rating'] = df['Rating'].fillna(df['Rating'].mean())   # 결측치를 평균값으로 대체 ( mean() : 평균 구하기 )

df.loc[index, 'Rating']   # index 확인
```

```python
# ffill : 앞의 값으로 뒤의 결측치를 대체

import numpy as np

s = pd.Series([1, np.nan, np.nan,2, np.nan, 3])
s.fillna(method="ffill")
```

```python
# dropna()
# 결측치가 포함된 행/열 삭제

import pandas as pd
import numpy as np

df = pd.read_csv('./datasets/bookings.csv')
df = df.dropna(subset='Review', axis=0)
df['Review'].isna().sum()   # 결측치가 모두 대체되었는지 확인
```

```python
# quantile : 해당 행/열의 분위수에 해당하는 값 반환
# lower, higher, midpoint, nearest, liner

quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]
for idx in quantile:
    q = df['Total_Review'].quantile(idx, interpolation='lower')
    print(f'quantile({idx}) is {q}')
```

```python
quantile = [0, 0.2, 0.4, 0.6, 0.8, 1]
for idx in quantile:
    q = df['Total_Review'].quantile(idx, interpolation='nearest')
    print(f'quantile({idx}) is {q}')
```

```python
# groupby

import seaborn as sns
df = sns.load_dataset('titanic')

df.groupby('sex')['survived'].mean()
```

```python
# groupby
# 항목을 sex로 분류 후 각각의 분류된 항목을 다시 class로 분류 / 각각의 항목에 대한 survived의 평균 출력

import seaborn as sns
df = sns.load_dataset('titanic')

df.groupby(['sex', 'class'])['survived'].mean()
```

```python
# groupby
# 항목을 sex로 분류 후 각각의 분류된 항목을 다시 class로 분류 / 각각의 항목에 대한 survived의 평균과 빈도수를 출력

import seaborn as sns
df = sns.load_dataset('titanic')

df.groupby(['sex', 'class'])['survived'].agg(['mean', 'count'])
```

```python
# groupby
# 항목을 sex로 분류 후 각각의 분류된 항목을 다시 class로 분류 / 각각의 항목에 대한 survived의 평균과 age의 최대값을 출력

import seaborn as sns
df = sns.load_dataset('titanic')

df.groupby(['sex','class'])[['survived','age']].agg({'survived':'mean', 'age':'max'})
```

```python
# groupby

import seaborn as sns
df = sns.load_dataset('titanic')

def get_IQR(data):
    _3rd = data.quantile(.75)
    _1st = data.quantile(.25)
    return (np.abs(_3rd - _1st) * 1.5)
    
df.groupby(['sex','class'])['age'].apply(get_IQR)
    
```

```python
# species로 분류한 뒤 항목들에 대한 평균값 확인

import seaborn as sns
df = sns.load_dataset('penguins')

df.groupby('species')[[ 'bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g' ]].mean()
```

```python
# groupby : group으로 분류하여 각각의 항목들에 대한 value의 합을 출력

import pandas as pd

df = pd.DataFrame(
    {'group':['A','A','A','B','B'],
     'value':[1, 1, 1, 10, 10]}
)

df.groupby('group')['value'].sum()
```

```python
# groupby 인자로 열 이름 이외 다른 형태의 데이터 전달

import pandas as pd

df = pd.DataFrame(
    {'group':['A','A','A','B','B'],
     'value':[1, 1, 1, 10, 10]}
)

df.groupby([0,0,1,1,1])['value'].sum()
```

```python
# groupby 인자로 열 이름 이외 다른 형태의 데이터 전달

import pandas as pd

df = pd.DataFrame(
    {'group':['A','A','A','B','B'],
     'value':[1, 1, 1, 10, 10]}
)

s = pd.Series([False, False, True, True, True])
df.groupby(s)['value'].sum()
```
